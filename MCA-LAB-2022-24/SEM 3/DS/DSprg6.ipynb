```python
df = pd.read_csv("Cancer_DS.csv")
```

```python
df.dtypes
```

```python
df.drop(["id","Unnamed: 32"], axis = 1, inplace = True)
```

```python
df.columns
```

```python
M = df[df.diagnosis == "M"]
B = df[df.diagnosis == "B"]

plt.scatter(M.radius_mean,M.texture_mean, color = "blue", label = "Malignant", alpha = 0.3)
plt.scatter(B.radius_mean,B.texture_mean, color = "red", label = "Benign", alpha = 0.3)

plt.xlabel("radius_mean")
plt.ylabel("texture_mean")

plt.legend()
plt.show()
```

```python
plt.scatter(M.smoothness_mean,M.compactness_mean, color = "blue", label = "Malignant", alpha = 0.3)
plt.scatter(B.smoothness_mean,B.compactness_mean, color = "red", label = "Benign", alpha = 0.3)

plt.xlabel("smoothness_mean")
plt.ylabel("compactness_mean")

plt.legend()
plt.show()
```

```python
feature=df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']]
labels=df[['diagnosis']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(feature,labels,test_size=0.2,random_state=42)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
```

```python
print(f"Accuracy{accuracy:.2f}")
```
